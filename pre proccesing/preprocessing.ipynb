```python
import re

def preprocess_name(name):
    # Loại bỏ dấu ( và )
    name = re.sub(r'[()]', '', name)
    # Loại bỏ các từ MRS, MR, MISS
    name = re.sub(r'\b(MRS|MR|MISS)\b', '', name)
    # Loại bỏ các dấu \t, \ và / và dấu xuống dòng \n, thay bằng dấu cách trắng
    name = re.sub(r'[\t\\/\\\n]', ' ', name)
    # Lấy 2 chữ (2 chuỗi kí tự có dấu cách trắng ở giữa)
    name = re.sub(r'\s{2,}', ' ', name)
    # Lấy 2 chữ
    name = ' '.join(name.split()[:2])
    return name.strip()  # Loại bỏ khoảng trắng đầu và cuối chuỗi

# Áp dụng hàm tiền xử lý vào cột "Tên khách"
try_model_NameToNat['Tên khách'] = try_model_NameToNat['Tên khách'].apply(preprocess_name)
```
