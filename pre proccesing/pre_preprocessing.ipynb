```python
import pandas as pd
dich_vu3 = pd.read_excel(r'C:\Users\trand\Documents\study\university\23_24_ki2\CAP1\dieu_hanh\dich_vu\dich_vu2.xlsx')
```

```python
dich_vu3 = dich_vu3.groupby(['Đối tác', 'Dịch vụ'])['Số đơn đặt'].sum().reset_index(name='Số đơn đặt')
dich_vu3
```

```python
new_table = dich_vu3.pivot_table(index='Dịch vụ', columns='Đối tác', values='Số đơn đặt', aggfunc='sum')
new_table.reset_index(inplace=False)
new_table.fillna(0, inplace=True)  # Điền các giá trị NaN bằng 0 nếu có
new_table
```

```python
new_table.dtypes
```

# tổng hợp các đối tác theo dịch vụ

```python
# Lấy danh sách tên cột đối tác
partners = new_table.columns.tolist()

# Tạo một danh sách để lưu các DataFrame tạm thời
dfs = []

# Duyệt qua từng dịch vụ
for service in new_table.index:
    # Lấy danh sách đối tác có giá trị lớn hơn 0
    selected_partners = [partner for partner in partners if new_table.loc[service, partner] > 0]
    # Tạo một DataFrame tạm thời
    temp_df = pd.DataFrame({'Dịch vụ': [service], 'Đối tác': [', '.join(selected_partners)]})
    # Thêm DataFrame tạm thời vào danh sách
    dfs.append(temp_df)

# Ghép các DataFrame tạm thời thành một DataFrame lớn
new_df = pd.concat(dfs, ignore_index=True)
new_df
```

```python
new_df.to_excel('tong_hop_dv.xlsx')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import re

# Sao chép DataFrame để tránh ảnh hưởng đến DataFrame gốc
df_23_copy = df_23.copy()

# Tạo biểu thức chính quy để tìm kiếm ngày tháng trong chuỗi
date_pattern = r'\d{1,2}/\d{1,2}/\d{4}'

# Lọc ra các giá trị hợp lệ từ cột 'K/H' và chuyển đổi thành datetime
df_23_copy['K/H'] = df_23_copy['K/H'].apply(lambda x: re.findall(date_pattern, str(x)))
df_23_copy['K/H'] = pd.to_datetime(df_23_copy['K/H'].str[0], errors='coerce')

# Loại bỏ các hàng có giá trị không hợp lệ
df_23_copy = df_23_copy.dropna(subset=['K/H'])

# Nhóm theo tháng và đếm số lượng mỗi tháng
monthly_count = df_23_copy.resample('M', on='K/H').size()

# Vẽ biểu đồ line chart
plt.figure(figsize=(10, 6))
monthly_count.plot(kind='line', marker='o', linestyle='-')
plt.title('Biểu đồ line chart với số lượng mỗi tháng')
plt.xlabel('Thời gian')
plt.ylabel('Số lượng')
plt.grid(True)
plt.show()
```

```python
import os
import pandas as pd

# Đường dẫn tới thư mục chứa các file Excel
folder_path = r'C:\Users\trand\Documents\study\university\23_24_ki2\CAP1\Sim\2023'

# Danh sách tên các file Excel trong thư mục
file_names = os.listdir(folder_path)

# Khởi tạo DataFrame trống để lưu trữ dữ liệu từ các file Excel
merged_df = pd.DataFrame()

# Đọc và merge dữ liệu từ các file Excel
for file_name in file_names:
    if file_name.endswith('.xlsx'):
        file_path = os.path.join(folder_path, file_name)
        df = pd.read_excel(file_path)
        merged_df = pd.concat([merged_df, df], ignore_index=True)

# Loại bỏ các dòng trùng lặp
merged_df.drop_duplicates(inplace=True)

# In ra số lượng dòng và cột sau khi merge và loại bỏ dòng trùng lặp
print("Số lượng dòng và cột sau khi merge và loại bỏ dòng trùng lặp:", merged_df.shape)

# Lưu DataFrame sau khi merge và loại bỏ dòng trùng lặp vào file Excel
merged_file_path = os.path.join(folder_path, 'merged_data_SIM23.xlsx')
merged_df.to_excel(merged_file_path, index=False)

print("Dữ liệu sau khi merge và loại bỏ dòng trùng lặp đã được lưu vào file:", merged_file_path)
```
